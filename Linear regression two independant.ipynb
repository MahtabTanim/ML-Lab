```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
Data = pd.read_csv("sample.csv")
Data
```

```python
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Air")
plt.scatter(Data.area,Data.air,c=Data.price, cmap='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Three Variables')
plt.colorbar(label='Z-axis')
```

```python
Lin = linear_model.LinearRegression()
Lin.fit(Data[['area', 'air']],Data.price)
Lin.coef_
Lin.intercept_
```

```python
X =[[4200,115],[4500,120],[5000,99]]
```

```python
Y = Lin.predict(X)
Y
```
