```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
Data = pd.read_csv("homeprices.csv")
Data
```

```python
Data.bedrooms.median()
```

```python
Data.bedrooms = Data.bedrooms.fillna(Data.bedrooms.median())
Data
```

```python
Lin = linear_model.LinearRegression()
```

```python
Lin.fit(Data[['area','bedrooms','age']],Data.price)
```

```python
Lin.coef_
```

```python
Lin.intercept_
```

```python
Lin.predict([[2600,3.0,20]])
```

```python
%matplotlib inline
plt.title('Linear Regression with Three Variables')
plt.xlabel("Area")
plt.xlabel("price")
plt.scatter(Data.area,Data.price)
```
