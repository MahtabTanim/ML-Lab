```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
areas = np.array([2600, 3000, 3200, 3600, 4000]).reshape(-1, 1)  # Reshape to 2D array
prices = np.array([550000, 565000, 610000, 680000, 725000])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(areas,prices,color='red',marker = '*')
```

```python
reg1 = linear_model.LinearRegression()
reg1.fit(areas,prices)
reg1.predict([[3300]])
```

```python
reg1.coef_
```

```python
reg1.intercept_
```

```python
df2 = pd.read_csv("areas.csv")
```

```python
predict2 =reg1.predict(df2)
predict2
```

```python
df2['price'] = predict2
```
